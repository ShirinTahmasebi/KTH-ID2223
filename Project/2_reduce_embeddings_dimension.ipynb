```python
RESULT_DATASET_PATH = '/content/drive/MyDrive/Embedding with GPT-3/output/result_Reviews.csv'
```

```python
engines = \
[
  'text-similarity-davinci-001',
  'text-similarity-ada-001',
  # 'text-similarity-babbage-001',
  # 'text-similarity-curie-001',
  # 'code-search-ada-code-001',
  # 'code-search-babbage-code-001'
]
```

```python
import pandas as pd

df = pd.read_csv(RESULT_DATASET_PATH, index_col=0)
```

```python
df
```

```python
import json
from sklearn.manifold import TSNE
import numpy as np

perplexity = 30
tsne_2d = TSNE(n_components=2, verbose=1, perplexity=perplexity, n_iter=3000, random_state=5)

for engine in engines:
  embedding_column_name = 'embedding_' + engine
  reduced_x_column_name = 'reduced_x_' + engine
  reduced_y_column_name = 'reduced_y_' + engine
  
  fetched_embedding_series = df[embedding_column_name].apply(lambda x: json.loads(x))

  fetched_embedding_arr = np.empty((0, len(fetched_embedding_series.iloc[0])), float)
  for item in fetched_embedding_series:
    fetched_embedding_arr = np.append(fetched_embedding_arr, np.array(item).reshape(1, -1), axis=0)
  
  reduced_embeddings = tsne_2d.fit_transform(fetched_embedding_arr)

  reduced_x = reduced_embeddings[:,0]
  reduced_y = reduced_embeddings[:,1]
  
  df[reduced_x_column_name] = reduced_x
  df[reduced_y_column_name] = reduced_y

  # break # Because other embeddings are null so far
```

```python
import os

def check_dir(file_name):
    directory = os.path.dirname(file_name)
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
check_dir('output/result_Reviews.csv')
df.to_csv('output/result_Reviews.csv')
```
